```python
import pytesseract
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
#for configuration
pytesseract.pytesseract.tesseract_cmd = ("C:\\Program Files\\Tesseract-OCR\\tesseract")
```

```python
img = cv2.imread("C:\\Users\\dell\\Desktop\\computer_vision\\images_of_project\\1632759271_pdf_application\\p3.png")
```

```python
plt.imshow(img)
```

```python
img2char = pytesseract.image_to_string(img)
```

```python
print(img2char)
```

```python
imgbox = pytesseract.image_to_boxes(img)
```

```python
print(imgbox)
```

```python
type(imgbox)
```

```python
imgH, imgW,_ = img.shape
```

```python
img.shape
```

```python
for boxes in imgbox.splitlines():
    boxes = boxes.split(' ')
    x,y, w,h = int(boxes[1]),int(boxes[2]),int(boxes[3]),int(boxes[4])
    cv2.rectangle(img, (x,imgH-y),(w,imgH-h),(0,0,255),3)
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY))
```
